```python
#The code can be integrated into Django by creating a view to handle the form input, processing it with the model in a separate module,

import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, TextIteratorStreamer
from threading import Thread


model_name = "LGAI-EXAONE/EXAONE-Deep-2.4B"
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.bfloat16,
    trust_remote_code=True
)
model = model.to("cuda")


tokenizer = AutoTokenizer.from_pretrained(model_name)

patient_name = "Harsh"  
patient_age = 35
patient_gender = "Male"
#PATIENT INFORMATION TO BE RECIEVED FROM DATABASE DURING FINAL INTEGRATION

def generate_output(prompt):
    input_ids = tokenizer(prompt, return_tensors="pt").input_ids.to("cuda")
    streamer = TextIteratorStreamer(tokenizer)
    thread = Thread(target=model.generate, kwargs=dict(
        input_ids=input_ids,
        eos_token_id=tokenizer.eos_token_id,
        max_new_tokens=10240,
        do_sample=True,
        temperature=0.6,
        top_p=0.95,
        streamer=streamer
    ))
    thread.start()
    output_text = ""
    for text in streamer:
        output_text += text
        print(text, end="", flush=True)
    print()
    return output_text


print("Welcome to the Neurology Assistant Chatbot.\n")

print("Chatbot: What symptoms are you experiencing?")
symptoms = input("You: ")

print("\nChatbot: How long have you been experiencing these symptoms?")
duration = input("You: ")

print("\nChatbot: Do you have any previous medical conditions? (yes/no)")
conditions = input("You: ")

context = f"{symptoms}; {duration}; {conditions}"

prompt = f"""
You are a medical assistant. Based on the following patient responses, generate a detailed medical report with clear sections (Patient History, Clinical Findings, Assessment, Recommendations) and include a final line with the preliminary diagnosis. 

Patient Details:
Name: {patient_name}
Age: {patient_age}
Gender: {patient_gender}

Patient Responses:
Symptoms: {symptoms}
Duration: {duration}
Previous Conditions: {conditions}

The final line must strictly follow this format:
Preliminary Diagnosis: <diagnosis>

JUST A REPORT AND A DIAGNOSIS AT THE END NOTHING ELSE PLEASE
"""

print("\nGenerating report and preliminary diagnosis...\n")
generate_output(prompt)
```
